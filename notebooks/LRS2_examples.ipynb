```python
from lrs2multi import LRS2Multi
from ..lrs2object import LRS2Object
import glob
import os.path as op
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

def get_scifiles_from_folder(folders, object_name=None):
    filenames = []
    for folder in folders:
        all_names = sorted(glob.glob(op.join(folder, 'multi*.fits')))
        for filename in all_names:
            filenames.append(filename)
    smaller_list = []
    for filename in filenames:
        f = fits.open(filename)
        if ('uv' in filename) or ('orange' in filename):
            side = '056'
        else:
            side = '066'
        name = f[0].header['OBJECT']
        slot = name.split('_')[-2]
        if side != slot:
            continue
        if object_name is None:
            smaller_list.append(filename)
        else:
            if object_name.lower() in name.lower():
                smaller_list.append(filename)
    return smaller_list

def get_scifiles_from_folder_from_pos(folders, object_name=None, exposure_min=None):
    filenames = []
    for folder in folders:
        all_names = sorted(glob.glob(op.join(folder, 'multi*.fits')))
        for filename in all_names:
            filenames.append(filename)
    smaller_list = []
    for filename in filenames:
        f = fits.open(filename)
        if ('uv' in filename) or ('orange' in filename):
            side = '056'
        else:
            side = '066'
        name = f[0].header['OBJECT']
        slot = name.split('_')[-2]
        name = 'J%s%s' % (''.join(f[0].header['QRA'].split(':')), ''.join(f[0].header['QDEC'].split(':')))
        if side != slot:
            continue
        if exposure_min is not None:
            if f[0].header['EXPTIME'] < exposure_min:
                continue
        if object_name is None:
            smaller_list.append(filename)
        else:
            if object_name.lower() in name.lower():
                smaller_list.append(filename)
    return smaller_list
    
sns.set_context('poster')
sns.set_style('ticks')
plt.rcParams["font.family"] = "Times New Roman"

colors = plt.rcParams['axes.prop_cycle'].by_key()['color'] * 10
def_wave = np.arange(3650., 10500, 0.7)
```

```python
folders = ['/Users/gregz/cure/LRS2Multi/for_basmah/']
objectname1 = 'J032906.05+303039.20'
objectname2 = 'J042930.08+243955.05'
objectname3 = 'J032851.26+311739.30'
objectname4 = 'J032927.60+312109.82'
objectname5 = 'J032904.95+312038.40'
objectname6 = 'J032859.23+312032.50'
objectname7 = 'J182940.21+001513.10'
objectname8 = 'J182957.66+011304.60'
objectname9 = 'J033022.45+313240.50'
objectname10 = 'J041858.00+281223.34'
objectname11 = 'J032924.09+311957.60'
objectname12 = 'J034513.82+321210.00'

objectname = objectname4

filenames = get_scifiles_from_folder_from_pos(folders, object_name=objectname, exposure_min=300)

detwave = 6563.
wave_window= 150
extraction_radius = 3.5
sky_radius = 3.
object_radius = 2.
red_detect_channel = 'red'
blue_detect_channel = 'orange'
func = np.nanmean

LRS2 = LRS2Object(filenames, detwave=detwave, wave_window=wave_window,
                  red_detect_channel=red_detect_channel,
                  blue_detect_channel=blue_detect_channel)

LRS2.subtract_sky(func=np.nansum, local=False, pca=False, sky_radius=sky_radius, obj_radius=object_radius)

for key in LRS2.sides.keys():
    for L in LRS2.sides[key]:
        if (L.channel == blue_detect_channel) or (L.channel == red_detect_channel):
            L.plot_image(radius=extraction_radius, func=func, attr='skysub', quick_skysub=False)

LRS2.extract_spectrum(radius=extraction_radius)
LRS2.smooth_resolution(redkernel=1.8)
LRS2.rectify(def_wave)
LRS2.normalize(detwave=detwave, wave_window=wave_window, func=func)
LRS2.get_astrometry()
#LRS2.make_cube(def_wave)
LRS2.calculate_sn()
LRS2.combine_spectra()
LRS2.write_combined_spectrum()


wlim = [4500, 9900]
fig, ax = plt.subplots(1, 1, figsize=(14, 6), sharex=True, gridspec_kw={'hspace':0.0})
ax.set_position([0.14, 0.14, 0.8, 0.8])

LRS2.plot_spectrum()
plt.plot(LRS2.spec1D.spectral_axis.value, LRS2.spec1D.flux.value, 'k-', lw=0.5)

ax.tick_params(axis='both', which='both', direction='in')
ax.tick_params(axis='y', which='both', left=True, right=True)
ax.tick_params(axis='x', which='both', bottom=True, top=True)
ax.tick_params(axis='both', which='major', length=8, width=2)
ax.tick_params(axis='both', which='minor', length=5, width=1)
ax.minorticks_on()
plt.xlim([wlim[0], wlim[1]])
ylim = [np.nanpercentile(LRS2.spec1D.flux.value, 0.2), np.nanpercentile(LRS2.spec1D.flux.value, 99.8)]
ran = ylim[1] - ylim[0]
plt.ylim([ylim[0] - ran*0.1, ylim[0] + ran*1.5])
ax.set_ylabel(r'F$_{\lambda}$ (erg / s / cm$^2$ / $\mathrm{\AA}$)')
ax.set_xlabel(r'Wavelength ($\mathrm{\AA}$)')
keys = list(LRS2.sides.keys())
plt.savefig('%s_spectrum_plot.png' % objectname, dpi=150)
```
